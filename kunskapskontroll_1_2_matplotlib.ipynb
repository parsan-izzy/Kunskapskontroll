### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x value from -4 to 4
x = np.linspace(-4, 4, 100)

#Calculate y value
y = x**2 

# Plot the function using pyplot
plt.plot(x, y)

# Set title and axis lables
plt.title("Plot of y = x^2")
plt.xlabel("x")
plt.ylabel("y = x^2")

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Create a figure with two subplots
fig, (scatter_plot, bar_plot) = plt.subplots(1, 2, figsize = (12, 5))



# First subplot: Scatter plot
scatter_plot.scatter(random_data_x, random_data_y, alpha = 0.4, color = 'red')
scatter_plot.set_title('Scatter Plot')
scatter_plot.set_xlabel('X-axis')
scatter_plot.set_ylabel('Y-axis')

# Data for bar plot
fruit_data = {'grapes' : 22, 'apple' : 8, 'orange' : 15, 'lemon' : 20, 'lime' : 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Second subplot: Bar plot
bar_plot.bar(names, values, color = 'green')
bar_plot.set_title('Bar Plot')
bar_plot.set_xlabel('Fruits')
bar_plot.set_ylabel('Quantity')

# Adjust layout to prevent overlap
plt.tight_layout()

plt.show()
```
